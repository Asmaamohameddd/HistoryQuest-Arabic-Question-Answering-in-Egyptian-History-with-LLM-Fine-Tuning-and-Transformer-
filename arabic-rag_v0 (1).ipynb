# load Book

```python
from langchain_community.document_loaders import PyPDFLoader

book_path = r'/kaggle/input/test1234r/.pdf'

loader = PyPDFLoader(book_path)
loader
```

```python
pages = loader.load_and_split()
```

```python
print(f"this book contains {len(pages)} pages")
print(pages[1].page_content)
```

```python
# clean pages
for page in pages:
    content = page.page_content
    len_before = len(content)
    content = content.replace("\n", "").replace("\t", "").replace("-", "")
    len_after = len(content)
    print(len_before, len_after, f"saved {len_before-len_after}")
    page.page_content = content
```

# Split Book

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size=1000,
    chunk_overlap=500,
    length_function=len,
    is_separator_regex=False,
)
```

```python
splitted_docs = splitter.transform_documents(pages)
```

```python
print(len(splitted_docs))
splitted_docs[2]
```

# Convert from `Documents` into plain text

```python
texts = []
for doc in splitted_docs:
    texts.append(doc.page_content)
```

# Embedding model

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': False}
embedding_model_id = "asafaya/bert-large-arabic"
hf = HuggingFaceEmbeddings(
    model_name = embedding_model_id,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

# Database Creation

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(splitted_docs, hf)
```

# Create Retriever

```python
db.save_local("faiss_db")
```

```python
retriever = db.as_retriever()
```
